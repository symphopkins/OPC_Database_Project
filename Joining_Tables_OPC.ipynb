**Part 3: Write a join statement that returns all customers who placed an order that shipped to Ohio from the Sacramento warehouse. Only return the following columns:**

- **Customer ID**
- **Order Number**
- **Order Date**
- **Order Total**

````sql
--The order tax location for Ohio customers is 20.
SELECT sac.cus_id
      ,o.ord_id 
      ,o.ord_date
      ,o.order_tot
FROM 
(SELECT cwo.cus_id
      ,cwo.warehouse_id
      ,w.warehouse_name
FROM dsci_504.customerwarehouseorders AS cwo
INNER JOIN dsci_504.warehouses AS w
ON cwo.warehouse_id = w.warehouse_id
WHERE w.warehouse_name = 'Sacramento') AS sac
LEFT JOIN dsci_504.orders AS o
ON sac.cus_id = o.cus_id 
WHERE o.ord_tax_loc = 20;
````

````sql
SELECT q1.cus_id
      ,q1.cus_last_name
      ,q1.cus_first_name
      ,q1.cus_app_cd 
      ,q1.warehouse_id 
      ,q1.ord_date 
      ,q1.ord_id
FROM
(SELECT c.cus_id
      ,c.cus_last_name
      ,c.cus_first_name
      ,c.cus_app_cd 
      ,r.warehouse_id 
      ,o.ord_date 
      ,o.ord_id
      ,r.tot_ret_item_cnt
      ,s.sup_name
FROM dsci_504.customers AS c 
INNER JOIN dsci_504.racs AS r 
ON c.cus_id = r.cus_id 
INNER JOIN dsci_504.orders AS o 
ON r.ord_id = o.ord_id 
INNER JOIN dsci_504.products AS p 
ON r.prod_id = p.prod_id
INNER JOIN dsci_504.suppliers AS s 
ON CAST(p.prod_manufacturer AS INT) = s.sup_id
WHERE r.tot_ret_item_cnt > 0) AS q1
WHERE q1.sup_name = 'Commencal'
ORDER BY q1.cus_app_cd DESC
        ,q1.ord_date DESC;
````

````sql
SELECT o.ord_tax_loc
      ,p.prod_manufacturer
      ,p.prod_name
      ,SUM(o.order_tot)
FROM dsci_504.orders AS o 
INNER JOIN dsci_504.products AS p
ON o.prod_id = p.prod_id 
INNER JOIN dsci_504.suppliers AS s 
ON CAST(p.prod_manufacturer AS INT) = s.sup_id
WHERE s.sup_name = 'Yeti' OR s.sup_name = 'Santa Cruz'
GROUP BY ROLLUP (o.ord_tax_loc, p.prod_manufacturer, p.prod_name)
ORDER BY o.ord_tax_loc;
````

**Part 6A: Now write a query to return all customers whose orders shipped from a warehouse outside of their region.  It will take some creativity to complete this query. Only return the Customer ID, Order ID, and Customer Appreciation Level of the customers whose orders where shipped from a region outside of their own.**

```
Tip: You will need a CASE statement to complete this query. Be sure to use the customerwarehouseorders table to aid you. Use the below logic for your state-to-region CASE:
```
```
mid-west = ME, VT, NH, MA, RI, CT, NY, PA, NJ, DE, MD, VA, WV, KY, OH, IN, MI, WI, IL, MO, KS, IA, MN, ND, SD, NE
```
```
south = NC, TN, AR, OK, TX, NM, LA, MS, AL, GA, SC, FL
```
```
west = MT, WY, CO, AZ, UT, ID, WA, OR, CA, NV, AK, HI
```

````sql
SELECT q2.cus_id
      ,q2.ord_id
      ,q2.cus_app_cd
FROM
(SELECT
    CASE 
        WHEN state IN ('ME', 'VT', 'NH', 'MA', 'RI', 'CT', 'NY', 'PA', 'NJ', 'DE', 'MD', 'VA', 'WV', 'KY', 'OH', 'IN', 'MI', 'WI', 'IL', 'MO', 'KS', 'IA', 'MN', 'ND', 'SD', 'NE') THEN 'Midwest'
        WHEN state IN ('NC', 'TN', 'AR', 'OK', 'TX', 'NM', 'LA', 'MS', 'AL', 'GA', 'SC', 'FL') THEN 'South'
        WHEN state IN ('MT', 'WY', 'CO', 'AZ', 'UT', 'ID', 'WA', 'OR', 'CA', 'NV', 'AK', 'HI') THEN 'West'
        ELSE 'Undetermined'
    END AS cus_region
          ,q1.*
FROM
(SELECT o.cus_id
      ,c.cus_app_cd
      ,s.state
      ,o.ord_id
      ,w.warehouse_name
      ,w.region
FROM dsci_504.orders AS o 
INNER JOIN dsci_504.warehouses AS w 
ON o.warehouse_id = w.warehouse_id 
INNER JOIN dsci_504.states AS s 
ON s.state_id = CAST(o.ord_tax_loc AS INTEGER)
INNER JOIN dsci_504.customers AS c
ON o.cus_id = c.cus_id) AS q1) AS q2
WHERE q2.cus_region != q2.region
ORDER BY q2.cus_app_cd;
````
