**Part 1: Write a query that finds the total number of sales for the year 2015. Your query cannot use any aggregates. All calculations should be done using the data/time functions in this week's learning. Your output should be in two columns-- year\_sales and num\_of\_sales.**

```sql
SELECT DISTINCT(EXTRACT(YEAR FROM o.ord_date)) AS year_sales
      ,COUNT(ord_id) OVER (PARTITION BY EXTRACT(YEAR FROM o.ord_date)) AS num_of_sales
FROM dsci_504.orders AS o
WHERE EXTRACT(YEAR FROM o.ord_date) = 2015;
```

**Find the monthly sales for the same year use the same output as above.**

```sql
SELECT *
FROM
(SELECT DISTINCT(TO_CHAR(o.ord_date, 'YYYY-MM')) as year_sales
,COUNT(ord_id) OVER (PARTITION BY TO_CHAR(o.ord_date, 'YYYY-MM')) AS num_of_sales
FROM dsci_504.orders AS o) AS q1
WHERE q1.year_sales BETWEEN '2015-01' AND '2015-12'

```

**Part 3: Perform a query to find all customers with orders fulfilled through the Dallas warehouse. This should be a simple query with no joins. Order your results by the total value of each customer's order.**

```sql
--Dallas warehouse code is 2.
SELECT *
FROM  
(SELECT o.cus_id
      ,o.warehouse_id
      ,o.order_tot
      ,SUM(o.order_tot) OVER(PARTITION BY o.warehouse_id ORDER BY o.order_tot) AS cus_order_tot
FROM dsci_504.orders AS o) AS q1
WHERE q1.warehouse_id = 2;
```

**Write a query to calculate the average bike price, the sum of all bikes sold, and the difference between the product price and sale price of every bike purchase. You can include any other columns you feel necessary.**

```sql
SELECT p.prod_id 
      ,p.prod_name 
      ,p.prod_price
      ,o.order_tot
      ,o.order_tot - p.prod_price AS sales_diff
      ,AVG(p.prod_price) OVER() as avg_bike_price
      ,SUM(o.order_tot) OVER(PARTITION BY p.prod_name) AS tot_bike_sales
FROM dsci_504.products AS p
INNER JOIN dsci_504.orders AS o
ON p.prod_id = o.prod_id;
```

- **What is one anomaly detected in the query results?**
    - There is only one order where there is a sales difference that is negative (-580.07). This may have occurred because of a sale, but it is interesting to see that it is the only order where the order total is significantly less than the product price.
